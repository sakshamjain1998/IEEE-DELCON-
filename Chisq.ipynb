# CHI-sq testing - H/A

## Method 1

```python
def method_1_chisq(col_name):
  col_name_HA_crosstab=pd.crosstab(df[col_name], df['H/A'])
  col_name_WA_crosstab=pd.crosstab(df[col_name], df['W/A'])
  col_name_WH_crosstab=pd.crosstab(df[col_name], df['W/H'])
  col_name_Bmi_crosstab=pd.crosstab(df[col_name], df['Bmi'])
  display(col_name_HA_crosstab)
  display(col_name_WH_crosstab)
  display(col_name_WA_crosstab)
  display(col_name_Bmi_crosstab)

  print('Begin chi sq testing')

  from scipy import stats
  print('For H/A')
  (chi2_col_name_HA,p_col_name_HA,dof,_)=stats.chi2_contingency([col_name_HA_crosstab.iloc[0].values, col_name_HA_crosstab.iloc[1].values])
  print('chi2 for H/A:' ,chi2_col_name_HA)
  print('pvalue for H/A:' ,p_col_name_HA)
  print('For W/A')
  (chi2_col_name_WA,p_col_name_WA,dof,_)=stats.chi2_contingency([col_name_WA_crosstab.iloc[0].values, col_name_WA_crosstab.iloc[1].values])
  print('chi2 for W/A:' ,chi2_col_name_WA)
  print('pvalue for W/A:' ,p_col_name_WA)
  print('For WH')
  (chi2_col_name_WH,p_col_name_WH,dof,_)=stats.chi2_contingency([col_name_WH_crosstab.iloc[0].values, col_name_WH_crosstab.iloc[1].values])
  print('chi2 for W/H:' ,chi2_col_name_WH)
  print('pvalue for W/H:' ,p_col_name_WH)
  print('For Bmi')
  (chi2_col_name_Bmi,p_col_name_Bmi,dof,_)=stats.chi2_contingency([col_name_Bmi_crosstab.iloc[0].values, col_name_Bmi_crosstab.iloc[1].values])
  print('chi2 for Bmi:' ,chi2_col_name_Bmi)
  print('pvalue for Bmi:' ,p_col_name_Bmi)
```

## Method 2

```python
def method_2_chisq(col_name):

  x_dummy=pd.get_dummies(df[col_name])
  y1_dummy=pd.get_dummies(df['H/A'])
  y2_dummy=pd.get_dummies(df['W/A'])
  y3_dummy=pd.get_dummies(df['W/H'])
  y4_dummy=pd.get_dummies(df['Bmi'])
  kbc=SelectKBest(score_func=chi2, k='all')

  print('Begin Chi sq testing')
  print("  ")
  print("  ")
  print("  ")
  print('for H/A')
  kbc.fit_transform(x_dummy, y1_dummy)
  ch_scores=kbc.scores_
  p_values=kbc.pvalues_
  columns=list(x_dummy)
  print("  ")
  print('Features               ', 'ch_scores                    ', 'Pvalues  ')
  print("----------         ------------------------            ----------- ")
  for i in range(0, len(columns)):
    f1="%4.2f" % ch_scores[i]
    p1="%2.4f" % p_values[i]
    print('  ', columns[i].ljust(12),f1.rjust(12),"   ",p1.rjust(30))

  print("  ")
  print("  ")
  print("  ")
  print('for W/A')
  kbc.fit_transform(x_dummy, y2_dummy)
  ch_scores=kbc.scores_
  p_values=kbc.pvalues_
  columns=list(x_dummy)
  print("  ")
  print('Features               ', 'ch_scores                    ', 'Pvalues  ')
  print("----------         ------------------------            ----------- ")
  for i in range(0, len(columns)):
    f1="%4.2f" % ch_scores[i]
    p1="%2.4f" % p_values[i]
    print('  ', columns[i].ljust(12),f1.rjust(12),"   ",p1.rjust(30))
  
  print("  ")
  print("  ")
  print("  ")
  print('for W/H')
  kbc.fit_transform(x_dummy, y3_dummy)
  ch_scores=kbc.scores_
  p_values=kbc.pvalues_
  columns=list(x_dummy)
  print("  ")
  print('Features               ', 'ch_scores                    ', 'Pvalues  ')
  print("----------         ------------------------            ----------- ")
  for i in range(0, len(columns)):
    f1="%4.2f" % ch_scores[i]
    p1="%2.4f" % p_values[i]
    print('  ', columns[i].ljust(12),f1.rjust(12),"   ",p1.rjust(30))
  
  print("  ")
  print("  ")
  print("  ")
  print('for Bmi')
  kbc.fit_transform(x_dummy, y4_dummy)
  ch_scores=kbc.scores_
  p_values=kbc.pvalues_
  columns=list(x_dummy)
  print("  ")
  print('Features                       ', 'ch_scores                    ', 'Pvalues  ')
  print("----------                      ------------------------            ----------- ")
  for i in range(0, len(columns)):
    f1="%4.2f" % ch_scores[i]
    p1="%2.4f" % p_values[i]
    print('  ', columns[i].ljust(12),f1.rjust(12),"   ",p1.rjust(30))
```
